```python
def set_count_enquiry_recency(data, cust_enquiry, n):
    latest_enquiry_date = max(cust_enquiry["enquiry_dt"])
    latest_enquiry_window_start = latest_enquiry_date - pd.Timedelta(days=n)
    d = (cust_enquiry[["customer_no", "enquiry_dt"]]
         [cust_enquiry["enquiry_dt"] > latest_enquiry_window_start]
         .groupby("customer_no").agg("count")
        ).rename(columns={"enquiry_dt": f"count_enquiry_recency_{n}"})


    data[f"count_enquiry_recency_{n}"] = np.zeros(data.shape[0], dtype=int)
    data[f"count_enquiry_recency_{n}"].loc[d.index] = d[f"count_enquiry_recency_{n}"]
    
    
    
```

```python
import collections

def get_max_count(x):
    if len(x) == 0:
        return -1
    else:
        y = collections.Counter(x).most_common(1)[0][0]
        if np.isnan(y):
            return -1
        else:
            return y

    
# get_max_count([1, 2, 3, 2, 2])
# get_max_count([np.nan])


    
data["max_freq_enquiry"] = cust_enquiry[["customer_no", "enq_purpose"]].groupby("customer_no").agg(list)["enq_purpose"].apply(get_max_count)
data["max_freq_enquiry"].dtype


data["max_freq_enquiry"] = data["max_freq_enquiry"].astype(int)
```

```python
cibil_codes = {
"STD": -1, #"Standard (STD): Payments being made within 90 days",

"SMA": -2, #"Special account created for reporting Standard Accounts moving toward Sub-Standard",
"SUB": -3, #"Payments being made after 90 days",
"DBT": -4, #"The account has remained Sub-Standard for 12 months",
"LSS": -5, #"An account where loss has been identified and remains uncollectible",
"XXX": -6 #"on your DPD for a certain account implies that information for these months has not been reported to cibil"
}
    

def get_payment_history(s):
    if not isinstance(s, str) and np.isnan(s):
        return [cibil_codes['XXX']]
    s = s.replace('"', '')
    values = []
    
    for i in range(0, len(s), 3):
        value = s[i:i+3]
        try:
            value = int(value)
            values.append(value)
        except:
            if value in cibil_codes.keys():
                values.append(cibil_codes[value])
    return values

```

```python
cust_account["paymenthistory1"] = cust_account["paymenthistory1"].apply(get_payment_history)
cust_account["paymenthistory2"] = cust_account["paymenthistory2"].apply(get_payment_history)

# data["paymenthistory2"] = cust_account.apply(lambda x: get_payment_histories(x["paymenthistory1"]))
```

```python
cust_account["paymenthistory1"][0]

def get_0_29_dpd(x):
    bad = 0
    
    for i in x:
        if i == cibil_codes["STD"] or i == cibil_codes["XXX"]: # good accounts benefit of doubt to XXX -6 -1 are good
            pass
        elif i < 0: # bad accounts because cibil codes are encoded as negative numbers
            bad = 1
        else:
            if i > 29:
                bad = 1
    good = 1 - bad            
    return good

def is_bad(i):
    bad = False
    if i == cibil_codes["STD"] or i == cibil_codes["XXX"]: # good accounts benefit of doubt to XXX -6 -1 are good
        pass
    elif i < 0:
        bad = True
    else:
        if i > 29:
            bad = True
    return bad
 
    
get_0_29_dpd(cust_account["paymenthistory1"][0])

def get_num_months_passed_before_first_30dpd(x):
    """ The smallest number of months passed before first 30+ dpd appeared for each account. """
    count = 0
    
    for i in x:
        if not is_bad(i):
            count += 1
        else:
            break
    return count
    
    
get_num_months_passed_before_first_30dpd(cust_account["paymenthistory1"][11])

cust_account["0_29dpd1"] = cust_account["paymenthistory1"].apply(get_0_29_dpd)
cust_account["0_29dpd2"] = cust_account["paymenthistory2"].apply(get_0_29_dpd)

cust_account["num_passed_before_first_30dpd1"] = cust_account["paymenthistory1"].apply(get_num_months_passed_before_first_30dpd)
cust_account["num_passed_before_first_30dpd2"] = cust_account["paymenthistory2"].apply(get_num_months_passed_before_first_30dpd)

d = cust_account.groupby("customer_no").agg({"0_29dpd1": ["sum", "count"],
                                         "0_29dpd2": ["sum", "count"],
                                         
                                         "num_passed_before_first_30dpd1": "min",
                                         "num_passed_before_first_30dpd2": "min",                                         
                                        })

x = (d["0_29dpd1"]["sum"] + d["0_29dpd2"]["sum"]) / (d["0_29dpd2"]["count"] + d["0_29dpd2"]["count"])
data["payment_history_avg_dpd_0_29_bucket"] = x


```
